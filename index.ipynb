```python
travels_data = travel_data.copy()
```

```python
travels_data.drop_duplicates()
```

```python
def remove_space(text):
    if type(text) == str:
        return text.strip()
    return text
```

```python
travels_data['Travel End Date'] = travels_data['Travel End Date'].apply(remove_space)
```

```python
travels_data['Travel Start Date'] = pd.to_datetime(travels_data['Travel Start Date'])
travels_data['Travel End Date'] = pd.to_datetime(travels_data['Travel End Date'])
```

```python
#Defines funnctions to extract state and city
def extract_state(text):
    result = text.split(',')[1]
    return result

def extract_city(text):
    result = text.split(',')[0]
    return result

```

```python
#travels_data['Employee_first_name'] = travels_data['Employee'].apply(extract_state)
#travels_data['Employee_last_name'] = travels_data['Employee'].apply(extract_city)
```

```python
#Extracted state and city from address
travels_data['State'] = travels_data['Destination'].apply(extract_state)
travels_data['City'] = travels_data['Destination'].apply(extract_city)
```

```python
#drop columns
travels_data = travels_data.drop(columns=['city', 'state'])
```
