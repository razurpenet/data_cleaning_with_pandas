```python
def remove_space(text):
    if type(text) == str:
        return text.strip()
    return text
```

```python
travels_data['Travel End Date'] = travels_data['Travel End Date'].apply(remove_space)
```

```python
travels_data['Travel Start Date'] = pd.to_datetime(travels_data['Travel Start Date'])
travels_data['Travel End Date'] = pd.to_datetime(travels_data['Travel End Date'])
```

```python
#Defines funnctions to extract state and city
def extract_state(text):
    result = text.split(',')[1]
    return result

def extract_city(text):
    result = text.split(',')[0]
    return result

```

```python
#Extract state and city
travels_data['state'] = travels_data['Destination'].apply(extract_state)
travels_data['city'] = travels_data['Destination'].apply(extract_city)
```

```python
#Defines functions to extract first_name and last_name
def split_first_name(text):
    result = text.split(' ')[0]
    return  result

def split_last_name(text):
    result = text.split(' ')[-1]
    return  result
```

```python
#Extracted last_name and first_name from Employee
travels_data['first_name'] = travels_data['Employee'].apply(split_first_name)
travels_data['last_name'] = travels_data['Employee'].apply(split_last_name)
```
