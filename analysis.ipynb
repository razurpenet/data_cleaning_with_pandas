```python
travels_data = pd.read_csv('cleaned_travel_dataset.csv')
travels_data = travels_data.drop('Unnamed: 0', axis=1)
```

```python
employees_expenses = travels_data[['Employee', 'Actual Total Expenses']]
```

```python
employee_expense = employees_expenses.groupby(by='Employee').sum()
```

```python
employees_df = employee_df.drop_duplicates(subset = 'name')
```

```python
employees_expense = employee_expense.reset_index() # reset index to pandas default because groupby make the groupby column default index
```

```python
#join the two dataframe to obtain results

results = employees_expense.merge(employees_df, how='inner', left_on='Employee', right_on='name')
results
```
